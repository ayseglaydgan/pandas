```python
import pandas as pd
import numpy as np

df = pd.read_csv('country_vaccination_stats.csv')

def find_min(country):
    return df.loc[df['country'] == country]['daily_vaccinations'].min()

def find_null(country):
    if(df.loc[df['country'] == country, 'daily_vaccinations'].isnull):
        print('No data for ' + country )
        return df.loc[df['country'] == country]['daily_vaccinations'].fillna(find_min(country), inplace=True)
       
def main(): 
    countries = df.country.unique()
    tempAllDailyVaccinationsFrame : any
    firstTime = True
    for country in countries:
        dataFrameVaccine = df.loc[df['country'] == country]['daily_vaccinations']
        dataFrameVaccine.fillna(find_min(country), inplace=True)
        if firstTime:
            tempAllDailyVaccinationsFrame = dataFrameVaccine
            firstTime = False
        else:
            tempAllDailyVaccinationsFrame = tempAllDailyVaccinationsFrame.append(dataFrameVaccine)


    df.daily_vaccinations = tempAllDailyVaccinationsFrame 
    df.fillna(0, inplace=True)   

main()
df.to_csv('response.csv', index=False)





```
